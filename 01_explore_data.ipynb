```python
df = spark.table("default.raw_retail_source")
print(f"Rows: {df.count():,}  Cols: {len(df.columns)}")
df.printSchema()
display(df.limit(10))
```

```python
from pyspark.sql.functions import col, sum as _sum

null_counts = df.select([
    _sum(col(c).isNull().cast("int")).alias(c)
    for c in df.columns
])
display(null_counts)
```

```python
from pyspark.sql.functions import *

# Count cancellation invoices (start with "C")
cancels = df.filter(col("InvoiceNo").startswith("C")).count()
neg_qty  = df.filter(col("Quantity") <= 0).count()
zero_prc = df.filter(col("UnitPrice") <= 0).count()
null_cust = df.filter(col("CustomerID").isNull()).count()

print(f"Cancellation invoices : {cancels:,}")
print(f"Negative quantity rows : {neg_qty:,}")
print(f"Zero/negative price   : {zero_prc:,}")
print(f"Null CustomerID       : {null_cust:,}")
```

```python
display(df.groupBy("Country").count().orderBy(desc("count")).limit(10))

df.select(
    min("InvoiceDate").alias("earliest"),
    max("InvoiceDate").alias("latest")
).show()
```
